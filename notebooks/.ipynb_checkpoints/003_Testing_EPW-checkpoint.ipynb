```python
import pandas as pd
import matplotlib.pyplot as plt
from iertools.read import read_sql
from dateutil.parser import parse
```

```python
f = "../data/002_processed/3marzo2018.csv"
tmx = pd.read_csv(f,index_col=0,parse_dates=True)
tmx.columns
```

```python
f = "../idf/cubo.sql"
idf = read_sql(f,alias=True)
eplus = idf.data
eplus.columns
```

```python
var = 'TECHO:Surface Outside Face Incident Solar Radiation Rate per Area (W/m2)'
omega = 'Environment:Site Solar Hour Angle (deg)'
```

```python

fig, ax  = plt.subplots(3,1,sharex=True)
f1 = parse("2018-03-4")
f2 = f1 + pd.Timedelta("1D")


ax[0].plot(tmx.WS,"ko",label="esolmet")
ax[0].plot(eplus.ws,"r.",label="Eplus")


ax[1].plot(tmx.To,"ko",label="Ig esolmet")
ax[1].plot(eplus.To,"r.",label="To eplus")

ax[2].plot(tmx.WD,"ko",label="WD esolmet")
ax[2].plot(eplus.wd,"r.",label="WD eplus")

for grafica in ax:
    grafica.set_xlim(f1,f2)
    grafica.legend()
    grafica.grid()
```

```python
fig, ax  = plt.subplots(figsize=(10,6))
f1 = parse("2018-03-4 05:00")
f2 = f1 + pd.Timedelta("15H")

ax.plot(eplus[Ig],"ro",label="Ig E+",alpha=0.3)
ax.plot(tmx.Ig,"r.")
ax.plot(eplus.Ib, "go",label="Ib E+",alpha=0.3)
ax.plot(tmx.Ib,   "g.")
ax.plot(eplus.Id, "bo",label="Id E+",alpha=0.3)
ax.plot(tmx.Id,   "b.")



ax.set_xlim(f1,f2)
ax.legend()
ax.grid()
```
