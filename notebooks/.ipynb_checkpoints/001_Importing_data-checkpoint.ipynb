```python
import pandas as pd
import matplotlib.pyplot as plt
import math
from datetime import datetime, timedelta
```

```python

def ecuacion_tiempo_duffie(dia_del_anio):
    B = 360 * (dia_del_anio - 81) / 365
    B = math.radians(B)
    EoT_min = 9.87 * math.sin(2*B) - 7.53 * math.cos(B) - 1.5 * math.sin(B)
    return EoT_min

def hora_solar(fecha_y_hora):
    # Longitud de Temixco y Meridiano Estándar (Hora Estándar del Centro)
    longitud_temixco = -99.23
    meridiano_estandar = -90

    # Diferencia de longitud a tiempo
    delta_longitud = longitud_temixco - meridiano_estandar
    correccion_longitud = timedelta(minutes=delta_longitud * 4)  # 4 minutos por grado

    # Corrección de la ecuación de tiempo
    dia_del_anio = (fecha_y_hora - datetime(fecha_y_hora.year, 1, 1)).days + 1
    EoT = ecuacion_tiempo_duffie(dia_del_anio)
    correccion_et = timedelta(minutes=EoT)

    # Sumar las correcciones a la hora estándar
    hora_solar = fecha_y_hora + correccion_longitud + correccion_et

    return hora_solar

# Ejemplo de uso
fecha_y_hora_actual = datetime.now()

hora_solar_resultado = hora_solar(fecha_y_hora_actual)
print(f"Hora Estándar: {fecha_y_hora_actual.time()}")
print(f"Hora Solar: {hora_solar_resultado.time()}")
```

```python
f = "../data/001_raw/003.xls"
esolmet = pd.read_excel(f,skiprows=[0,1,3,4],
                        index_col=0,
                       parse_dates=True)
esolmet = esolmet.loc["2018-03"].resample("10Min").mean()
nombres = {'Directa':"Ib",
           'Global':"Ig",
           "Difusa":'Id',
           'Temperatura':"To",
           'Humedad':"HR",
           'Viento Vel.':"WS", 
           'Presiòn Bar.':"P"}
esolmet.rename(columns=nombres,inplace=True)
del esolmet["UV"]
esolmet.index = esolmet.index.to_series().apply(hora_solar)
esolmet.columns
```

```python
esolmet.plot(subplots=True);
```

```python
f = "../data/001_raw/2018-2019-VIENTO-TMIX-RUOA.txt"
ruoa = pd.read_csv(f,
                   skiprows=4,
                   delimiter="\t",
                   encoding="ANSI",
                   usecols=[0,5],
                   names=["time","WD"]
                  )
ruoa.time = pd.to_datetime(ruoa.time,
                           format="%d/%m/%Y %H:%M")
ruoa.set_index("time",inplace=True)
ruoa = ruoa.loc["2018-03"]
ruoa = ruoa.resample("10Min").mean()
ruoa.index = ruoa.index.to_series().apply(hora_solar)
ruoa.columns
```

```python
marzo = pd.concat([esolmet,ruoa],axis=1)
marzo = marzo.loc["2018-03-04":"2018-03-10"]
marzo.plot(subplots=True);
```

```python
marzo.to_csv("../data/002_processed/3marzo2018.csv")
```
